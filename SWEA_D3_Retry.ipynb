```python
#12741
T = int(input())
answer = list()
for test_case in range(1,1+T):
    a,b,c,d = list(map(int, input().split()))

    start = max(a,c)
    end = min(b,d)
    period = end - start
    if period > 0:
        answer.append(period)
    else:
        answer.append(0)
    
for i in range(len(answer)):
    print(f'#{i+1} {answer[i]}')
```

```python
import sys
sys.stdin = open('input.txt', 'r')

def recursive(numstr):
    if len(numstr) <= 1:
        return numstr
    else:
        numlst = list(map(int, list(numstr)))
        total = 0
        for i in numlst:
            total += i
        return recursive(str(total))

T = int(input())
answer = list()
for test_case in range(1, T + 1):
    n = input()
    answer.append(recursive(n))

for i in range(len(answer)):
    print(f'#{i+1} {answer[i]}')
```

```python

T = int(input())
for test_case in range(1, T + 1):
    lst = list(map(str, list(input())))

    n = len(lst)
    maximum = int(''.join(lst))
    minimum = int(''.join(lst))

    for i in range(n):
        for j in range(i+1,n):
            temp = lst[:]
            temp[i], temp[j] = temp[j], temp[i]
            if temp[0] == '0':
                continue
            maximum = max(int(''.join(temp)), maximum)
            minimum = min(int(''.join(temp)), minimum)

    print(f'#{test_case} {minimum} {maximum}')
```

```python
#4798
T = int(input())
for test_case in range(1, T + 1):
    lst = list(map(int, list(input())))
    baksuNow = 0
    hired = 0
    for i in range(len(lst)):
        if lst[i] > 0:
            if baksuNow < i:
                hired += (i - baksuNow)
                baksuNow = i + lst[i]
            else:
                baksuNow += lst[i]
    print(f'#{test_case} {hired}')
```

```python
T = int(input())
for test_case in range(1,1+T):
    n = int(input())
    k = 2
    origLen = len(str(n))
    nlst = [0] * 10
    orig = list(str(n))
    for i in orig:
        nlst[int(i)] += 1
    

    while True:
        flag = True
        now = str(n * k)
        k += 1
        if len(now) > origLen:
            print(f'#{test_case} impossible')
            break
        comp = list(str(now))
        compLst = [0] * 10
        for i in comp:
            compLst[int(i)] += 1

        for i in range(10):
            if nlst[i] < compLst[i]:
                flag = False
                break
            
        if flag:
            print(f'#{test_case} possible')
            break
```
