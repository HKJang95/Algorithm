```python
# 1984

T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    # ///////////////////////////////////////////////////////////////////////////////////
    cases = list(map(int, input().split()))
    maxVal = max(cases)
    minVal = min(cases)
    allAdded = sum(cases)
    avg = (allAdded - maxVal - minVal) / 8
    result = '#' + str(test_case) + ' ' + str(round(avg))
    print(result)
```

```python
# 1961
def ninety(lst, N):
    newLst = [[0 for x in range(N)] for y in range(N)]
    y = N - 1
    for i in range(N):
        for j in range(N):
            newLst[j][y] = lst[i][j]
        y -= 1
    return newLst

def mapper(lst, N):
    tomaplst = [''] * N
    for i in range(N):
        for j in range(N):
            tomaplst[i] += str(lst[i][j])
    return tomaplst


T = int(input())
for test_case in range(1, T + 1):
    N = int(input())
    lst = list()
    for i in range(N):
        lst.append(list(map(int, input().split())))
    ninetyLst = ninety(lst, N)
    ninetySecondLst = ninety(ninetyLst, N)
    ninetyThirdLst = ninety(ninetySecondLst, N)

    first = mapper(ninetyLst, N)
    second = mapper(ninetySecondLst, N)
    third = mapper(ninetyThirdLst, N)
    
    print(f'#{test_case}')
    for i in range(N):
        print(first[i], end=" ")
        print(second[i], end=" ")
        print(third[i])
```

```python
# 1983
T = int(input())
for test_case in range(1, T + 1):
    gradeStr = ['A+', 'A0', 'A-', 'B+', 'B0', 'B-', 'C+', 'C0', 'C-', 'D0']
    line = list(map(int, input().split()))

    students = line[0]
    target = line[1]

    scores = dict()
    grade = dict()
    scoreRatio = students // 10

    for i in range(1,students+1):
        line = list(map(int, input().split()))
        mid = line[0] * 0.35
        final = line[1] * 0.45
        assignment = line[2] * 0.2
        scores[i] = mid + final + assignment
    
    scores = sorted(scores.items(), key=lambda x:x[1], reverse=True)
    
    count = 0
    for i in scores:
        grade[i[0]] = gradeStr[0]
        if count < scoreRatio - 1:
            count += 1
        else:
            gradeStr.pop(0)
            count = 0
        
    print(f'#{test_case} {grade[target]}')
```

```python
def bunhae(number, target):
    current = target
    count = 0
    while True:
        if number % current > 0:
            return count
        else:
            current *= target
            count += 1

numLst = [2,3,5,7,11]
T = int(input())
for test_case in range(1, T + 1):
    countLst = list()
    num = int(input())
    for i in numLst:
        countLst.append(bunhae(num, i))
    
    print(f'#{test_case}', end=" ")
    for i in countLst:
        print(f'{i}', end=" ")
    print('')

    
```

```python
#1940
import sys
sys.stdin = open('input.txt', 'r')

T = int(input())
for test_case in range(1, T + 1):
    cmdCount = int(input())
    speed = 0
    dist = 0
    for i in range(cmdCount):
        line = list(map(int,input().split())) # 1 : 가속 / 2 : 감속
        if len(line) < 2:
            dist += speed
            continue
        else:
            cmd = line[0]
            vel = line[1]
        if cmd == 1:
            speed += vel
        else:
            speed -= vel
            if speed < 0:
                speed = 0
        dist += speed
    print(f'#{test_case} {dist}')
```
