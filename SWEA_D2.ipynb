```python
# 1984

T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    # ///////////////////////////////////////////////////////////////////////////////////
    cases = list(map(int, input().split()))
    maxVal = max(cases)
    minVal = min(cases)
    allAdded = sum(cases)
    avg = (allAdded - maxVal - minVal) / 8
    result = '#' + str(test_case) + ' ' + str(round(avg))
    print(result)
```

```python
#2001
T = int(input())
 
for test_case in range(1, T + 1):
    line = list(map(int, input().split()))
    N = line[0]
    M = line[1]
    lst = [[0 for x in range(N)]for y in range(N)]
 
    for i in range(N):
        lineIn = list(map(int, input().split()))
        for j in range(N):
            lst[i][j] = lineIn[j]
    maxVal = 0
 
    for x in range(N - M + 1):
        for y in range(N - M + 1):
 
            currentVal = 0
            for i in range(M):
                for j in range(M):
                    currentVal += lst[x + i][y + j]
            maxVal = max(maxVal, currentVal)
    print(f'#{test_case} {maxVal}')
```

```python
#1284
T = int(input())
for test_case in range(1, T + 1):
    line = list(map(int, input().split()))
    P = line[0]
    Q = line[1]
    R = line[2]
    S = line[3]
    W = line[4]

    priceA = W * P
    
    if R < W:
        basicB = R
        overB = W - R
        priceB = Q + (overB * S)
    else:
        priceB = Q
    
    minPrice = min(priceA, priceB)
    print(f'#{test_case} {minPrice}')
```

```python
#2005
T = int(input())
for test_case in range(1, T + 1):
    height = int(input())

    lst = [[0 for x in range(height)] for y in range(height)]

    for i in range(height):
        lst[i][0] = 1
        lst[i][i] = 1

    for i in range(2, height):
        for j in range(1, i):
            lst[i][j] = lst[i-1][j-1] + lst[i-1][j]

    print(f'#{test_case}')
    for i in range(height):
        for j in range(height):
            if lst[i][j] > 0:
                print(lst[i][j], end=" ")
        print('')
```

```python
# 1970
T = int(input())

for test_case in range(1, T + 1):
    money = int(input())
    countlst = [0] * 8
    moneylst = [50000,10000,5000,1000,500,100,50,10]

    i = 0
    for i in range(8):
        countlst[i] = money // moneylst[i]
        money -= countlst[i] * moneylst[i]

    print(f'#{test_case}')
    for i in countlst:
        print(i, end=" ")
    print('')
```
