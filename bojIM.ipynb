```python
#2669
lst = [[0 for x in range(101)] for y in range(101)]

for i in range(4):
    x1,y1,x2,y2 = list(map(int, input().split()))

    for x in range(x1,x2):
        for y in range(y1,y2):
            lst[x][y] = 1

total = 0
for i in lst:
    total += sum(i)
print(total)
```

```python
#2605

import sys
sys.stdin = open('input.txt', 'r')

n = int(input())

lst = list(map(int, input().split()))
res = list()

for i in range(1,n+1):
    pos = lst.pop(0)
    front = res[:len(res) - pos]
    front.append(i)
    back = res[len(res) - pos:]
    res = front + back
for i in res:
    print(i, end=" ")

```

```python
#13300
import sys
sys.stdin = open('input.txt', 'r')

students, limit = list(map(int, input().split()))
boy = [0] * 7
girl = [0] * 7

for i in range(students):
    sex, grade = list(map(int, input().split()))

    if sex == 0:
        girl[grade] += 1
    else:
        boy[grade] += 1

rooms = 0
for i in range(1,7):
    rooms += boy[i] // limit
    if boy[i] % limit > 0:
        rooms += 1
    rooms += girl[i] // limit
    if girl[i] % limit > 0:
        rooms += 1
print(rooms)
```

```python
#2628

import sys
sys.stdin = open('input.txt', 'r')

paperx, papery = list(map(int, input().split()))
cut = int(input())
cutx = [0]
cuty = [0]
for i in range(cut):
    cmd, pos = list(map(int, input().split()))

    if cmd == 1:
        cutx.append(pos)
    else:
        cuty.append(pos)

cutx.append(paperx)
cuty.append(papery)

cutx.sort()
cuty.sort()
maxx = 0
maxy = 0

for i in range(1,len(cutx)):
    maxx = max(maxx, cutx[i] - cutx[i-1])
for i in range(1,len(cuty)):
    maxy = max(maxy, cuty[i] - cuty[i-1])

print(maxx*maxy)
```

```python
#2559
import sys
sys.stdin = open('input.txt', 'r')

n,k = list(map(int, input().split()))
lst = list(map(int, input().split()))
now = 0
for i in range(k):
    now += lst[i]
maximum = now

for i in range(1,n-k+1):
    now -= lst[i-1]
    now += lst[i+k-1]
    maximum = max(now, maximum)
print(maximum)
```

```python
#2116
import sys
sys.stdin = open('input.txt', 'r')

def ceiling(index):
    if index == 0:
        return 5
    if index == 1:
        return 3
    if index == 2:
        return 4
    if index == 3:
        return 1
    if index == 4:
        return 2
    if index == 5:
        return 0

dices = int(input())

lst = [[0 for x in range(6)] for y in range(dices)]
for i in range(dices):
    lst[i] = list(map(int, input().split()))

maximum = 0
for i in range(6):
    floor = i
    ceil = ceiling(floor)
    thismax = 0
    for j in range(dices):
        sidemax = 0 # 이번 주사위 최대값
        for k in range(6):
            if k == floor or k == ceil: # 바닥이나 천장은 무시
                continue
            else:
                sidemax = max(sidemax, lst[j][k]) # 옆면 최대값 갱신
        thismax += sidemax # 이번면 최대값 더해줌
        if j + 1 < dices: # 다음 주사위 있으면
            floor = lst[j+1].index(lst[j][ceil]) # 바닥 갱신
            ceil = ceiling(floor) # 천장 갱신
    maximum = max(thismax, maximum)
print(maximum)
```

```python
#2635
num = int(input())

divided = num // 2
if num % 2 > 0:
    divided += 1
maxlst = [num]
for i in range(divided, num+1):
    lst = [num,i]
    index = 1
    while True:
        now = lst[index-1] - lst[index]
        if now >= 0:
            lst.append(now)
        else:
            break
        index += 1
    if len(maxlst) < len(lst):
        maxlst = lst[:]
print(len(maxlst))
for i in maxlst:
    print(i,end=" ")
```
