```python
#13300
import sys
sys.stdin = open('input.txt', 'r')

students, limit = list(map(int, input().split()))
boy = [0] * 7
girl = [0] * 7

for i in range(students):
    sex, grade = list(map(int, input().split()))

    if sex == 0:
        girl[grade] += 1
    else:
        boy[grade] += 1

rooms = 0
for i in range(1,7):
    rooms += boy[i] // limit
    if boy[i] % limit > 0:
        rooms += 1
    rooms += girl[i] // limit
    if girl[i] % limit > 0:
        rooms += 1
print(rooms)
```

```python
#2628

import sys
sys.stdin = open('input.txt', 'r')

paperx, papery = list(map(int, input().split()))
cut = int(input())
cutx = [0]
cuty = [0]
for i in range(cut):
    cmd, pos = list(map(int, input().split()))

    if cmd == 1:
        cutx.append(pos)
    else:
        cuty.append(pos)

cutx.append(paperx)
cuty.append(papery)

cutx.sort()
cuty.sort()
maxx = 0
maxy = 0

for i in range(1,len(cutx)):
    maxx = max(maxx, cutx[i] - cutx[i-1])
for i in range(1,len(cuty)):
    maxy = max(maxy, cuty[i] - cuty[i-1])

print(maxx*maxy)
```

```python
#2559
import sys
sys.stdin = open('input.txt', 'r')

n,k = list(map(int, input().split()))
lst = list(map(int, input().split()))
now = 0
for i in range(k):
    now += lst[i]
maximum = now

for i in range(1,n-k+1):
    now -= lst[i-1]
    now += lst[i+k-1]
    maximum = max(now, maximum)
print(maximum)
```

```python
#2116
import sys
sys.stdin = open('input.txt', 'r')

def ceiling(index):
    if index == 0:
        return 5
    if index == 1:
        return 3
    if index == 2:
        return 4
    if index == 3:
        return 1
    if index == 4:
        return 2
    if index == 5:
        return 0

dices = int(input())

lst = [[0 for x in range(6)] for y in range(dices)]
for i in range(dices):
    lst[i] = list(map(int, input().split()))

maximum = 0
for i in range(6):
    floor = i
    ceil = ceiling(floor)
    thismax = 0
    for j in range(dices):
        sidemax = 0 # 이번 주사위 최대값
        for k in range(6):
            if k == floor or k == ceil: # 바닥이나 천장은 무시
                continue
            else:
                sidemax = max(sidemax, lst[j][k]) # 옆면 최대값 갱신
        thismax += sidemax # 이번면 최대값 더해줌
        if j + 1 < dices: # 다음 주사위 있으면
            floor = lst[j+1].index(lst[j][ceil]) # 바닥 갱신
            ceil = ceiling(floor) # 천장 갱신
    maximum = max(thismax, maximum)
print(maximum)
```

```python
#2635
num = int(input())

divided = num // 2
if num % 2 > 0:
    divided += 1
maxlst = [num]
for i in range(divided, num+1):
    lst = [num,i]
    index = 1
    while True:
        now = lst[index-1] - lst[index]
        if now >= 0:
            lst.append(now)
        else:
            break
        index += 1
    if len(maxlst) < len(lst):
        maxlst = lst[:]
print(len(maxlst))
for i in maxlst:
    print(i,end=" ")
```

```python
#2564
import sys
sys.stdin = open('input.txt', 'r')

x,y = list(map(int, input().split()))
n = int(input())

lst = list()
for i in range(n):
    dir,pos = list(map(int, input().split()))
    lst.append((dir,pos))

dir,pos = list(map(int, input().split())) # 동근위치

dist = 0

for i in lst:
    if i[0] == dir: # 같은 방향
        dist += abs(i[1] - pos)
        continue
    
    left = 0
    right = 0
    if dir == 1: # 동근이 북
        if i[0] == 2: # 가게 남
            left += pos # 왼쪽 끝
            left += y # 왼쪽 맨하단
            left += i[1] # 오른쪽으로 쭉감

            right += x - pos # 오른쪽 끝
            right += y # 오른쪽 맨하단
            right += x - i[1] # 왼쪽으로 쭉감
            dist += min(left,right)
            continue
        if i[0] == 3: # 가게 서
            dist += pos # 왼쪽 끝
            dist += i[1] # 왼쪽 아래로 내려감
            continue
        if i[0] == 4: # 가게 동
            dist += x - pos # 맨 오른쪽으로
            dist += i[1] # 오른쪽 아래로 내려감
            continue

    if dir == 2: # 동근이 남
        if i[0] == 1: # 가게 북
            left += pos # 왼쪽 끝
            left += y # 왼쪽 상단
            left += i[1] # 오른쪽으로 쭉감

            right += x - pos # 오른쪽 끝
            right += y # 오른쪽 상단
            right += x - i[1] # 왼쪽으로 쭉감
            dist += min(left,right)
            continue
        if i[0] == 3: # 가게 서
            dist += pos # 왼쪽 끝
            dist += y - i[1] # 왼쪽 위로 올라감
            continue
            
        if i[0] == 4: # 가게 동
            dist += x - pos # 맨 오른쪽으로
            dist += y - i[1] # 오른쪽 위로 올라감
            continue

    if dir == 3: #동근이 서
        if i[0] == 1: # 가게 북
            dist += pos # 맨 위로
            dist += i[1] # 오른쪽으로 쭉
            continue
        if i[0] == 2: # 가게 남
            dist += y - pos # 아래로 쭉
            dist += i[1] # 오른쪽으로 쭉
            continue
        if i[0] == 4: # 가게 동
            left += pos # 위로 쭉
            left += x # 오른쪽으로 쭉
            left += i[1] # 아래로
            
            right += y - pos # 아래로 쭉
            right += x # 오른쪽으로 쭉
            right += y - i[1] # 위로 쭉
            dist += min(right, left)
            continue

    if dir == 4: # 동근이 동
        if i[0] == 1: # 가게 북
            dist += pos # 위로 쭉
            dist += x - i[1] # 왼쪽으로 쭉
            continue
        if i[0] == 2: # 가게 남
            dist += y - pos # 아래로 쭉
            dist += x - i[1] # 왼쪽으로 쭉
            continue
        if i[0] == 3: # 가게 서
            left += pos # 위로 쭉
            left += x # 왼쪽으로 쭉
            left += i[1] # 아래로 쭉

            right += y - pos # 아래로 쭉
            right += x # 왼쪽으로 쭉
            right += y - i[1] # 위로 쭉
            dist = min(right, left)
            continue

print(dist)
```

```python
#2491
n = int(input())

lst = list(map(int, input().split()))

length = 1
maxlen = 1
for i in range(n-1): # 커지는거
    if lst[i] <= lst[i+1]:
        length += 1
        if i + 1 == n - 1:
            maxlen = max(length, maxlen)
    else:
        maxlen = max(maxlen, length)
        length = 1

length = 1
for i in range(n-1): # 작아지는거
    if lst[i] >= lst[i+1]:
        length += 1
        if i + 1 == n - 1:
            maxlen = max(length, maxlen)
    else:
        maxlen = max(maxlen, length)
        length = 1

print(maxlen)
```

```python
#2309
lst = list()
for i in range(9):
    lst.append(int(input()))
target = sum(lst) - 100
flag = False
for i in range(9):
    for j in range(i+1,9):
        if lst[i] + lst[j] == target:
            lst[i] = 0
            lst[j] = 0
            flag = True
            break
    if flag:
        break
lst.sort()
for i in lst:
    if i > 0:
        print(i)

```
